Melakukan tahapan pada soal 1 dengan menggunakan tensorflow

```python
import tensorflow as tf
```

```python
#buat variabel skalar x dan simpan suatu nilai skalar
x_tf = tf.Variable(25)
x_tf
```

```python
x_tf.shape
```

```python
#buat variabel skalar y dan simpan suatu nilai skalar
y_tf = tf.Variable(25)
y_tf
```

```python
#jumlahkan x+y
x_tf + y_tf
```

```python
#jumlahkan x+y
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
#jumlahkan x+y
tf_sum.numpy()
```

```python
type(tf_sum.numpy())
```

```python
tf_float = tf.Variable(25., dtype=tf.float16)
tf_float
```
