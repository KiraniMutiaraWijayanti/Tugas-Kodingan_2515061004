3. lakukan soal nomor 2 menggunakan pytorch dan tensorflow

lakukan transpose pada vector x dan simpan pada variabel x_t

```python
import torch
import tensorflow as tf
```

```python
x_pt = torch.tensor([20, 2, 0])
x_pt
```

```python
#Transpose x_pt dan simpan ke variabel x_t
x_t_pt = x_pt.T
x_t_pt
```

```python
#tampilkan ukuran vector x_t
x_t_pt.shape
```

lakukan transpose pada vector y dan simpan pada variabel y_t

```python
y_pt = torch.tensor([1, 2, 3])
y_pt
```

```python
#Transpose y_pt dan simpan ke variabel y_t
y_t_pt = y_pt.T
y_t_pt
```

```python
#tampilkan ukuran vector y_t
y_t_pt.shape
```

```python
x_tf = tf.Variable([3, 2, 1])
x_tf
```

```python
# Transpose menggunakan tf.transpose
x_tf_t = tf.transpose(x_tf)
print("Transpose TensorFlow shape:", x_tf_t.shape)
```
